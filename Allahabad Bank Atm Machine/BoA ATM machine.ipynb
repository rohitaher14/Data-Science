```python
''' Reading csv file 
Storing the same to Data frame and displying the top 10 rows from df'''
import numpy as np
import pandas as pd
import pyttsx3
engine = pyttsx3.init('espeak') 
bdf=pd.read_csv("atmdata.csv")
bdf.head(10)
```

```python
# Greet function will not retrun any value, just to display the welcome
def greet():
    print("Welcome to Allahabad Bank")
    engine.say("Welcome to Allahabad Bank")
    engine.runAndWait()    
```

```python
'''verifyuser is udf which will retrun value based on the authontecation of user
Passing 3 parameters and retruning 1 parameter.''' 
def verifyuser(a,b,c):
    #print('a',a, 'b',b,'c',c)
    if a==b:
        chance = 1
        while chance < 4 :
            engine.say("Enter your 4 digit pin for the Card Number.")
            engine.runAndWait()
            pin = int(input('Enter your 4 digit pin for the Card no.'))
            if pin == c:  #matching pin number
                print('Card verified successfully.') 
                verified = 'y' 
                break
            else: 
                print('Pls. enter the correct pin') 
                chance = chance +1 
                if chance == 4: 
                    print('your card is blocked, contact nearest Branch.')
                    engine.say("your card is blocked, contact nearest Branch.")
                    engine.runAndWait()
                    verified ='n' 
                    break
    else: 
        print('Entered Card number not available.') 
        verified ='n' 
    return(verified)        
```

```python
# startran is to display the option for the user and retruning option selected to main program. 
def startran(x):
    print('\n',x.capitalize()+',','Welcome to Allahabad Bank Transactions Menu') # this will capitalize the name of card holder
    engine.say("Welcome to Allahabad Bank Transactions Menu")
    engine.runAndWait()
    print('For Balance Press 1 \n')
    print('For Withdraw Press 2 \n')
    print('For Deposit Press 3 \n')
    print('For Change Pin Press 4 \n')
    option = int(input('What would you like to choose?'))
    return(option)
```

```python
''' This is main program for ATM machine. After calling greet function, 
it will ask for card number and validate if card exist in dataframe.
If card exist then it will get the details for card user from data frame.
It will authenticatet the by calling funciton verifyuser.
Post verifying the program will call the transaction process by proving options. 
Based on the optoin selected it will perform the task and display the result.'''

chance = 1
restart = 'y'
verified = 'n'
greet()

engine.say("Enter your Card Number.")
engine.runAndWait()
cardn = int(input('Enter your Card No.'))

if cardn in bdf.values:
    # retrieving card details including Pin
    cardd = bdf['card_no'] [bdf['card_no'] == cardn].values
    cardi = bdf.loc[bdf['card_no'] == cardn].index[0] #to identify the row matching card number
    pinx = bdf.iat[cardi,5] # to retrieve the pin of the card number
    # authenticating user
    verify = verifyuser(cardn,cardd,pinx)
    if verify == 'y':
        name = (bdf.iat[cardi,1])
        while restart not in ('n', 'NO', 'no', 'N'):
            option = startran(name)
            if option == 1:
                balance = bdf.iat[cardi,3]
                balance = "{:0,.2f}".format(float(balance)) # displaying amount in format
                print('Your Balance is Rs. ', balance, '\n')
                restart=input('Would you like to go back?')
                if restart in ('n', 'NO', 'no', 'N'):
                    print('Thank you',name.capitalize()+'.')
                    break
            elif option == 2:
                withdrawl = int(input('How much would you like to withdrawl? \n Enter Amount in 100s Rs. '))
                engine.say("Your transaction is processing, please wait")
                engine.runAndWait()
                if (withdrawl%100) == 0: # this is to verify the amount entered is in 100's 
                    balance = bdf.iat[cardi,3]
                    if balance < withdrawl: 
                        print('Balance amount is not sufficient \n')
                    else :
                        balance = balance - withdrawl 
                        bdf.loc[bdf['card_no'] == cardn, 'acc_bal'] = balance
                        print('Amount disburseed through ATM Rs.', withdrawl, 'Balance remaining is Rs.',balance, '\n')
                        engine.say("Kindly collect the cash from dispenser")
                        engine.runAndWait()
                else:
                    print('Incorrect amount entered, Amount should be in 100s')
                restart=input('Would you like to go back?')
                if restart in ('n', 'NO', 'no', 'N'):
                    print('Thank you',name.capitalize()+'.')
                    break
            elif option == 3:
                deposit = int(input('Kinldy inser the cash and enter the amount. \n Enter Amount in 100s Rs. '))
                if (deposit%100) == 0: # this is to verify the amount entered is in 100's 
                    balance = bdf.iat[cardi,3]
                    balance = balance + deposit 
                    bdf.loc[bdf['card_no'] == cardn, 'acc_bal'] = balance
                    print('You have deposited amount Rs.', deposit, 'Now Balance is Rs.',balance, '\n')
                else:
                    print('Incorrect amount entered, Amount should be in 100s')
                restart=input('Would you like to go back?')
                if restart in ('n', 'NO', 'no', 'N'):
                    print('Thank you',name.capitalize()+'.')
                    break
            elif option == 4:
                npin = int(input('New pin. \n'))
                cnpin = int(input('confirm pin. \n'))
                if npin == cnpin:
                    bdf.loc[bdf['card_no'] == cardn, 'pin'] = npin
                    print('Pin changed successfully.\n')
                    engine.say("Pin changed succesfully")
                    engine.runAndWait()
                else:
                    print('New pin and confirm pin entered are not matched.\n')
                restart=input('Would you like to go back?')
                if restart in ('n', 'NO', 'no', 'N'):
                    print('Thank you',name.capitalize()+'.')
                    break
            else:
                print('Please Enter a correct option. start over the process again \n')
    else:
        print('Your card is not verified.')
                
else:
    print('Card number enter is not correct. Kindly enter valid card number.')
                          
                
    
```
